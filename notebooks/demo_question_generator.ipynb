```python
# 3) Run the generator script (creates project/data/questions.csv)\nimport subprocess, sys, os\nscript = 'project/scripts/question_gen.py'\nif os.path.exists(script):\n    print('Running', script)\n    res = subprocess.run([sys.executable, script], capture_output=True, text=True)\n    print('--- STDOUT ---')\n    print(res.stdout)\n    print('--- STDERR ---')\n    print(res.stderr)\nelse:\n    raise FileNotFoundError(f'{script} not found — run this notebook from the repo root')\n
```

```python
# 4) Show generated questions (if present)\nimport pandas as pd, os\nout = 'project/data/questions.csv'\nif os.path.exists(out):\n    df = pd.read_csv(out)\n    display(df.head(20))\nelse:\n    print(out, 'not found — run the generator first')\n
```
